```python
import random

def computer_guess(x):
    low = 1
    high = x
    feedback = ''

    while feedback != "c":
        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low
        feedback = input( f'Is {guess} too high (H) low (L) or correct (c) ').lower()
        if feedback == 'h':
            high == guess - 1
        elif feedback == 'l':
            low = guess + 1
    print(f'yay! the Computer Guesses your Number . {guess} , Correctly')


computer_guess(10)
```
