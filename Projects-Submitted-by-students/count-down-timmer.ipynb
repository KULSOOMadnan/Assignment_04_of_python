```python
import time

    # Add welcome message
print("""
╔══════════════════════════════╗
║     Welcome to PyTimer!      ║
║  Your Personal Countdown     ║
╚══════════════════════════════╝
""")

print("\nWelcome! Let's set up your countdown timer.")

def count_down(t):

    print("\n⏰ Timer Started! ⏰\n")
    while t:

        # hours = 1 (because 3600 goes into 3725 one time)
        # rem = 125 (the remaining seconds after taking out 1 hour)
        hours, rem = divmod(t, 3600)
        # Second divmod to get minutes and seconds from remaining
        minutes, sec = divmod(rem, 60)
        timer_format = '⏳ {:02d}:{:02d}:{:02d}'.format(hours, minutes, sec)
        print(timer_format, end='\r')
        time.sleep(1)
        t -= 1
    print("\n�� Time's up! 🔔")


t = input("⌚ Enter the time in seconds: ")

count_down(int(t))
```
